```python
def extract_speech(path):
    sample = open(path, encoding = 'utf-8')
    sample = json.loads(sample)

    document = sample['document'][0]
    utterance = document['utterance']
    speakers = list(set(document['metadata']['speaker'][i]['id'] for i in range(len(document['metadata']['speaker']))))

    speaker_collect = collections.defaultdict(dict)
    for speaker in speakers:
        speaker_collect[speaker] = dict({'metadata': [], 'utterance': {'original_form': [], 'form': []}})

    # add metadata
    for speaker in speaker_collect:
        for metadata in sample['document'][0]['metadata']['speaker']:
            if speaker == metadata['id']:
                speaker_collect[speaker]['metadata'] = metadata

    # add utterances(origina_form, form)
    for speaker in speaker_collect:
        for utterance in utterance:
            if speaker == utterance['speaker_id']:
                speaker_collect[speaker]['utterance']['original_form'].append(f'{utterance["original_form"]}')
                speaker_collect[speaker]['utterance']['form'].append(f'{utterance["form"]}')

    return speaker_collect
```

```python
def extract_speech(path):
    sample = open(path, encoding = 'utf-8').read()
    sample = json.loads(sample)

    document = sample['document'][0]
    speakers = list(set(document['metadata']['speaker'][i]['id'] for i in range(len(document['metadata']['speaker']))))

    speaker_collect = collections.defaultdict(dict)
    for speaker in speakers:
        speaker_collect[speaker] = dict({'metadata': [], 'utterance': {'original_form': [], 'form': []}})

    for speaker in speaker_collect:
            # add metadata
        for metadata in document['metadata']['speaker']:
            if speaker == metadata['id']:
                speaker_collect[speaker]['metadata'] = metadata
            # add utterances(origina_form, form)
        for utterance in document['utterance']:
            if speaker == utterance['speaker_id']:
                speaker_collect[speaker]['utterance']['original_form'].append(f'{utterance["original_form"]}')
                speaker_collect[speaker]['utterance']['form'].append(f'{utterance["form"]}')

    return speaker_collect
```

```python
def extract(path):
    with open(path, encoding = 'utf8') as f:
        sample = json.loads(f.read())

    document = sample['document'][0]
    utterance = pd.DataFrame(document['utterance'])

    speakers = dict()
    for speaker in document['metadata']['speaker']:
        speakers[speaker['id']] = speaker['sex']

    utterance['count'] = pd.Series(map(lambda x: len(x.split()), utterance['form']))
    utterance['duration'] = np.array(utterance['end'], dtype = np.float32) - np.array(utterance['start'], dtype = np.float32)
    utterance['sex'] = [speakers[id] for id in utterance.speaker_id]
    utterance['sex'] = pd.Series(map(lambda x: 'female' if x == '여성' else 'male', utterance['sex']))
    utterance = utterance.drop(['original_form', 'start', 'end', 'note'], axis = 1)
    utterance = pd.get_dummies(utterance, prefix = 'sex', columns = ['sex'])

    return document, utterance
```
